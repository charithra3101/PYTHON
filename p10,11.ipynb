```python
#10.Write a Python program to demonstrate Timeseries analysis with Pandas.
import pandas as pd
df=pd.read_csv("AirPassengers.csv")
print(df.head())
print(df.tail())
df['Month']=pd.to_datetime(df['Month'],format='%Y-%m')
print(df.head())
df.index=df['Month']
del df['Month']
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(df)
plt.ylabel("Number of Passengers")

rolling_mean=df.rolling(7).mean()
rolling_std=df.rolling(7).std()

plt.plot(df,color="blue",label="Original Passengers Data")
```

```python
rolling_mean=df.rolling(7).mean()
rolling_std=df.rolling(7).std()
```

```python
plt.plot(df,color="blue",label="Original Passengers Data")
```

```python
plt.plot(df,color="blue",label="Original Passengers Data")
plt.plot(rolling_mean,color="red",label="Rolling Mean Passengers Number")
plt.plot(rolling_std,color="black",label="Rolling standard Deviation in Passengers Number")
plt.title("Passenger Time Series,Rolling Mean,Standard Deviation")
plt.legend(loc="best")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose=seasonal_decompose(df['#Passengers'],model='additive',period=7)
decompose.plot()
plt.show()
```

```python
#11.Write a Python program to demonstrate Data Visualization using Seaborn.
import seaborn as sns
import matplotlib.pyplot as plt

data=sns.load_dataset("iris")

sns.lineplot(x="sepal_length",y="sepal_width",data=data)

plt.xlim(5)
```

```python
plt.figure(figsize=(2,4))
```

```python
sns.set_context("paper")

plt.title('Title using Matplotlib Function')

sns.set_style("dark")

sns.despine()

plt.show()
```
